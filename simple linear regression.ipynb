IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import sklearn
```

IMPORT DATASET   URLIN GMAIL

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/cpluz-shrijayan/simple-Linear-Regression/main/Night%20Skill%20Training%20Dataset.csv")
dataset
```

SELECT X AND Y

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values
print(x)
print(y)
```

SPLIT DATASET INTO TRAIN AND TEST DATA

```python
from sklearn.model_selection import train_test_split
(x_train,x_test,y_train,y_test)=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

USE LINEAR REGRESSION TO TRAIN THE MODEL

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

PREDICT THE MODEL

```python
y_pred=lr.predict(x_test)
```

VISUALISE 

```python
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,lr.predict(x_train),color="blue")
plt.title("x_train vs predict_x_train")
plt.xlabel("x_train")
plt.ylabel("predict_x_train")
plt.show()
```

```python
plt.scatter(x_test,y_test,color="red")
plt.plot(x_test,lr.predict(x_test),color="blue")
plt.title("x_test vs predict_x_test")
plt.xlabel("x_test")
plt.ylabel("predict_x_test")
plt.show()
```
